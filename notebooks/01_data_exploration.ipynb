# Import necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set plot style

sns.set(style="whitegrid")



crops_df = pd.read_csv("../data/raw/crops_conditions.csv")  # Dataset containing crops & ideal climate conditions
domesticated_plants_df = pd.read_csv("../data/raw/domesticated_plants.csv")  # Dataset containing all crops data
locations_df = pd.read_csv("../data/raw/locations_climate.csv")  # Dataset containing locations & their climate data

# Display first few rows of each dataset
print("üü¢ Crops Dataset Preview:")
display(crops_df.head())

print("\nüü¢ Areas Dataset Preview:")
display(locations_df.head())

# Check dataset info
print("\nüîç Crops Dataset Info:")
print(crops_df.info())

print("\nüîç Areas Dataset Info:")
print(locations_df.info())
```

```python
# Check for missing values
print("\nüõë Missing Values in Crops Dataset:")
print(crops_df.isnull().sum())

print("\nüõë Missing Values in Areas Dataset:")
print(locations_df.isnull().sum())
```

```python
# Check for duplicate rows
print("\nüîÅ Duplicate Rows in Crops Dataset:", crops_df.duplicated().sum())
print("üîÅ Duplicate Rows in Areas Dataset:", locations_df.duplicated().sum())
```

```python
# Descriptive statistics
print("\nüìä Summary Statistics for Numerical Columns (Crops Dataset):")
print(crops_df.describe())

print("\nüìä Summary Statistics for Numerical Columns (Locations Dataset):")
locations_df.describe()
```

```python
# Standardizing column names (removes spaces and converts to lowercase)
locations_df.columns = locations_df.columns.str.lower().str.strip()

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(locations_df["temperature range (¬∞c)"], kde=True, ax=axes[0], color="red")
axes[0].set_title("Temperature Range Distribution")

sns.histplot(locations_df["annual precipitation (mm)"], kde=True, ax=axes[1], color="blue")
axes[1].set_title("Rainfall Distribution")

sns.histplot(locations_df["relative humidity (%)"], kde=True, ax=axes[2], color="green")
axes[2].set_title("Humidity Distribution")

plt.show()
```

```python
print(locations_df.dtypes)


# Define columns that should be numeric
numeric_columns = [
    "temperature range (¬∞c)", "light intensity (hours/day)", "relative humidity (%)",
    "annual precipitation (mm)", "altitude (m)", "day length (hours)"
]

# Convert these columns to numeric, forcing errors to NaN
for col in numeric_columns:
    locations_df[col] = pd.to_numeric(locations_df[col], errors="coerce")

# Extract only numeric columns from the DataFrame
numeric_data = locations_df[numeric_columns]  # Subset only numeric columns

# Drop rows with NaN values to avoid issues in correlation computation
numeric_data = numeric_data.dropna()

# Ensure numeric_data is not empty before plotting
if not numeric_data.empty:
    plt.figure(figsize=(8, 6))
    sns.heatmap(numeric_data.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
    plt.title("Correlation Matrix of Climate Conditions")
    plt.show()
else:
    print("No numeric data available for correlation.")
```

```python
# Identify numeric columns (ignoring categorical ones like 'country', 'region')
numeric_cols = [
    "temperature range (¬∞c)", "light intensity (hours/day)", "relative humidity (%)",
    "annual precipitation (mm)", "altitude (m)", "day length (hours)"
]

# Convert numeric columns (force errors to NaN)
for col in numeric_cols:
    locations_df[col] = pd.to_numeric(locations_df[col], errors="coerce")

# Extract numeric-only DataFrame
numeric_data = locations_df[numeric_cols]

# Drop rows where all values are NaN
numeric_data = numeric_data.dropna(how="all")

# Ensure we have data to plot
if not numeric_data.empty:
    plt.figure(figsize=(8, 6))
    sns.heatmap(numeric_data.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
    plt.title("Correlation Matrix of Climate Conditions")
    plt.show()
else:
    print("‚ö†Ô∏è No valid numeric data for correlation matrix.")
```
